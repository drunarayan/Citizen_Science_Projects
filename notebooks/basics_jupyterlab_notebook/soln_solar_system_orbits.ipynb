```python
# Import modules that contain functions we need
import pandas as pd # pandas is common for data science
import numpy as np # NumPy is used a lot in science
import math # we need the math class for pi
%matplotlib inline 
import matplotlib.pyplot as plt # MatPlotLib is the plotting standard
```

```python
# Read in the Github data that will be used for the calculations.
data1 = pd.read_csv("solar_system_orbits.csv")
data1
```

```python
def xval(angle):
    return a*np.cos(angle)

def yval(angle):
    return b*np.sin(angle)

## the linspace method defines the actual number points within the range
theta = np.linspace(0,2*np.pi, num=1000)
```

```python
plt.title("Inner Planetary Orbits in the Solar System!")
# Sun coordinates at the origin

#planets = ["Mercury", "Venus", "Earth", "Mars", "Ceres", "Pallas", 
#           "Jupiter", "Saturn", "Uranus", "Neptune", "Pluto",
#           "Vest", "Encke", "Halley", "Haumea", "Makemake", "Eris"]
#planets = ["Mercury", "Jupiter"]
rotate = 0;
for aPlanet in range(0,8):
    a = np.array(data1['Semimajor Axis (AU)'][aPlanet])
    P = np.array(data1['Period (y)'][aPlanet])
    e = np.array(data1['Eccentricity'][aPlanet])
    c=a*e
    b=(a**2-c**2)**0.5
    plt.xlabel("Distance AU")
    plt.ylabel("Distance AU");
    plt.gca().set_aspect('equal')
    plt.plot(xval(theta),yval(theta))
    #plt.scatter(u, v, marker ='+', color='black')
    #plt.scatter(+c, 0, marker ='o', color='gray')

    plt.scatter(xval(np.pi/4+rotate),yval(np.pi/4+rotate))
    plt.text(xval(np.pi/4+rotate),yval(np.pi/4+rotate),np.array(data1['Planet'][aPlanet]))
    rotate += 0.5

plt.scatter(0, 0, marker ='o', color='orange')    
plt.show()
```
